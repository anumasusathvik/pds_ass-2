```python
raw_data['Mileage'] = raw_data['Mileage'].str.extract('(\d+\.\d+)').astype(float)  # Extract and convert to float
raw_data['Engine'] = raw_data['Engine'].str.extract('(\d+)').astype(float)  # Extract and convert to float
raw_data['Power'] = raw_data['Power'].str.extract('(\d+\.\d+)').astype(float)  # Extract and convert to float
```

```python
import re
```

```python
def extract_numeric_value(text):
    if pd.notna(text):
        matches = re.findall(r'(\d+\.\d+|\d+)', str(text))
        if matches:
            return float(matches[0])
    return None
```

```python
#Task for Removeing units of Mileage, Engine, Power and extracting the Numerical values for the Coloumn
raw_data["Mileage"] = raw_data["Mileage"].apply(extract_numeric_value)
raw_data["Engine"] = raw_data["Engine"].apply(extract_numeric_value)
raw_data["Power"] = raw_data["Power"].apply(extract_numeric_value)
```
